```python
#Create a function to find out top 3 teams against whom a batsman has scored the most runs
def runs_scored(batsman_name):
    batsman = delivery[delivery['batsman'] == batsman_name]
    return batsman.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head(3)
```

```python
runs_scored('V Kohli')
```

```python
runs_scored('MS Dhoni')
```

#### Membership test

```python
#find out the most explosive death over batsman in the history of IPL
#strike rate = (Number of runs/Number of balls)*100
#Batsman should have faced atleast 200 balls in over 16-20
```

```python
death_over = delivery[delivery['over']>15]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
```

```python
all_batsman[x] #we have performed boolean indexing
```

```python
batsman_list = all_batsman[x].index.tolist()
```

```python
#we need runs scored by these 43 batsman and balls faced by them in the death overs
final = death_over[death_over['batsman'].isin(batsman_list)] #the isin() method needs a list input
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = runs/balls*100
sr.sort_values(ascending = False)
```

#### The Merge function

```python
#We want to fetch the list of orange cap holder for every season
#We will have to connect the two databases to get details about the season and runs scored by a batsman
match = pd.read_csv("matches.csv")
delivery = pd.read_csv("deliveries.csv")
```

```python
new = delivery.merge(match,left_on='match_id',right_on='id')
```

```python
new.shape
```

```python
(new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending = False).reset_index()
     .drop_duplicates(subset = 'season',keep='first').sort_values('season'))[['season','batsman']]
#we are grouping first based on season and then based on batsman
#we are getting a multi-index series that consists of 2 indices
#reset_index lets us convert our series output into a dataframe
#for more understanding execute parts of code from left to right
```

#### Pivot Tables in Pandas

```python
#We are going to show the total number of sixes hit by each team in every over of the match
mask = delivery['batsman_runs'] == 6
six = delivery[mask]
six.shape
```

```python
pt = six.pivot_table(index = 'over', columns = 'batting_team', values = 'batsman_runs', aggfunc = 'count')
pt
```

```python
#lets plot heatmap for the above pivot table
import seaborn as sns
sns.heatmap(pt)
```
